## DATA PROCESSING 

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
```

### 1 --> Defective Heart

### 0 --> Healthy Heart

```python
X = data.drop(columns='target', axis=1)
Y = data['target']
```

```python
print(X)
```

```python
print(Y)
```

### Splitting the Data into Training data & Test Data

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

## MODEL TRAINING

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
train_acc=[]
test_acc=[]

for k in range(1,31):
    model=KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train,y_train)
    # training time accuracy
    train_acc.append(model.score(x_train,y_train))
    # test time accuracy
    test_acc.append(model.score(x_test,y_test))
```

```python
train_acc
```

```python
plt.plot(range(1,31), train_acc,label="training accuracy")
plt.plot(range(1,31),test_acc,label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("K Neighbours")
plt.legend()
plt.savefig("KNN Comparison Models")
```

```python
knn_model=KNeighborsClassifier(n_neighbors=30)
knn_model.fit(x_train,y_train)
train_accuracy=knn_model.score(x_train,y_train)
test_accuracy=knn_model.score(x_test,y_test)
print("Accuracy in Training: ",train_accuracy)
print("Accuracy in testing: ", test_accuracy)
```

#### Limitations :

- Dataset Constraints: The performance of the models is highly dependent on the quality and quantity of the dataset used. - Ethical and Privacy Concerns: Using patient data for training machine learning models raises significant ethical and privacy concerns.
